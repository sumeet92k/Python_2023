# Plotting with matplotlib

```python
a = np.arange(10)
b = np.power(a, 2)
display(a)
display(b)
```

```python
plt.scatter(a, b, s=30, c='r', marker='*')
# plt.legend()
plt.show()
```

## Customizing scatterplots

```python
# plt.scatter(a, b, c='r', s=80, marker='*', label='squared')
plt.scatter(a, np.power(a, 1.5), label='raised to 1.5')
plt.scatter(a, b, c=b, s=80, marker='s', label='squared', cmap='rainbow')
plt.xlabel('number', fontsize=20)
plt.ylabel('powered', fontsize=20)
plt.title("Powers of numbers", fontsize=24)
plt.legend(fontsize=16)
plt.colorbar()
plt.show()
# plt.savefig("powered.png", dpi=300)
```

# Classes and objects
Classes used for organizing data. 
* Classes contain attributes (variables) and methods (functions)

Important properties/terms in classes:
* Class: template or structure to store data and functions
* Object: An *instance* of a class
* Instantiating: Creating an object of a class
* Constructor: Used to initialize the object
* Destructor: Used to destroy the object
* Encapsulation: All attributes and methods used by the class is contained inside. A class is stand alone
* Setters and Getters: For setting and fetching attributes
* Inheritance

## A simple class

```python
class Person():
    name=None
    age=None
    aadhaar=None
```

```python
p1 = Person()
```

```python
type(p1)
```

```python
p1.name = 'dhoni'
p1.age = 40
p1.aadhaar = 1234
```

```python
print(p1.name)
print(p1.age)
```

## Constructor \_\_init\_\_ method 

```python
class Person():
    def __init__(self, name=None, age=0, aadhaar=0):
        self.name = name # public
        self.age = age # public
        self.__aadhaar = aadhaar # private (hidden) from user
    
    def get_aadhaar(self): # getter function to fetch private (hidden) attribute
        print(self.__aadhaar)
```

## Attribute types
* Public attributes
* Protected and Private attributes are in Python for name sake. One can use '\_' prefix to inform user that the variable is not be accessed, but nothing is stopping the user from accessing it.
* '\_\_' prefix can be used to hide attributes from the user (make it protected). But the user can still access these using some tricks
* The above also applies to class methods

Getter and Setter functions to fetch and set variables

```python
p1 = Person('dhoni', age=40, aadhaar=1234)
```

```python
print(p1.name)
print(p1.age)
```

```python
print(p1.aadhaar)
```

```python
p1.get_aadhaar()
```
